下面通过yfinance提取中美互联网公司的股票数据，并做可视化分析。

```python
# 导入各类包
import matplotlib.pyplot as plt
import yfinance as yf  
%matplotlib inline

# 设置高清显示图表
%config InlineBackend.figure_format = 'svg'

# 设置中文字体
plt.rcParams['font.family']=['Adobe Heiti Std']
```

## 提取数据
考虑到公司的对比性，选取了中国互联网三巨头（BAT）阿里（股票代码：BABA），腾讯（0700.HK），百度（BIDU），以及美国互联网的三家巨无霸(FAG)：脸书（FB），亚马逊（AMZN），谷歌（GOOGL）。

获取数据需要设置起始日期和结束日期，我们要比较2018年以来这六家公司的股票走势，所以起始日期设置为‘2018-01-01’，结束日期为2019-08-27。

```python
# 设置要提取的股票代码为中国的BAT和美国的谷歌、脸书、亚马逊
tickers = ['0700.HK', 'BABA', 'BIDU', 'GOOGL', 'FB', 'AMZN']

# 设定起始日期和结束日期
start_date = '2018-01-01'
end_date = '2019-08-27'

# 提取雅虎财经数据
data = yf.download(tickers = tickers, start = start_date, end = end_date)
```

```python
# 查看前5行数据
data.head()
```

可以看到，数据的行索引是交易日日期，列索引有两层，第一层是当天股票的开盘价，最高价等变量，第二层列标题是股票代码。

由于我们需要比较的是2018年以来的股价走势，所以只需考虑收盘价。数据中有两种收盘，Close和Adj Close，分别代表不复权收盘价和前复权收盘价。股票分红、转股等情况下股价会除权，因此计前复权收盘价更能体现股价长期的真实变化（收益）。关于复权的概念，可以参考这篇[文章](https://www.zhihu.com/question/31004373)。

因此我们需要的数据在Adj Close列。

```python
# 提取前复权收盘价
df = data['Adj Close']

# 查看前5行
df.head()
```

此时只含有一级列索引，即股票代码。下面直接利用`dataframe.plot()`绘制股价图：

```python
df.plot()
```

根据提取的数据绘制出了这6家公司股价的走势，每家公司的每股价格差距较大，亚马逊和谷歌每股股价高于1000美金，其他几家股价数值低于500，并且腾讯的股价单位是港币。从上图直观上可以看到2018年以来亚马逊的股价是有明显上涨的，其他几家公司的股价走势要么持平于2018年初，要么走低。

为了比较各公司股价的涨跌幅，需要对数据进行处理，将该股所有的股价都除以该股起始日期（本文中的2018年1月1日）的股价，即将股票绝对值变成相对值，这样就能比较2018年以来这六家公司的股票收益。我们设定100为起始日期基准价，之后股价的波动都是相对100这个基准价而言的，这样也能忽略掉不同货币单位的影响。

另外，由于腾讯是在港股上市，港股和美股的休市日期不同，因此股价曲线会有几处中断，代表该日是休市日，该股无交易。对于休市日股价，可以认为休市日收盘价等于上一个交易日的收盘价，这样就可以使用`df.fillna()`通过前值填充休市日的股价，使股价曲线无间断。

```python
# 复制df的数据到df2
df2 = df.copy()

# 填充休市日股价为上一个交易日收盘价
df2.fillna(method='ffill', inplace=True)

# 使用for循环将每个公司的股价基于起始日收盘价折算，tickers为之前定义的公司股票代码组成的列表
# tickers = ['0700.HK', 'BABA', 'BIDU', 'GOOGL', 'FB', 'AMZN']
for ticker in tickers:
    
    # 该股全部收盘价除以起始日期收盘价，然后乘以100，股价按惯例保留2位小数
    df2[ticker] = round(df2[ticker] / df2[ticker].iloc[0] * 100, 2)

# 查看处理后的数据
df2.head()
```

每列的起始日期收盘价都变成了100，其他日期的股价直接反映了当日股票相对于起始日的收益率，这样就可以方便地作图比较2018年以来上述6家公司的股票收益。

```python
# 作图查看
df2.plot()
```

作图效果达到，但图表不够美观，图例直接挡住了股价线。可以再对图表做一些美化，添加标题，调整图例等。图例中使用的是股票代码，结合股票代码和公司名组成的字典，通过map函数将图例变成显示公司名，方便让不熟悉股票代码的读者看图。

```python
# 股票中文名
cn_name = ['腾讯', '阿里', '百度', '谷歌', '脸书', '亚马逊']

# 设置股票对应的颜色组
six_color = ['lightseagreen', 'lightsalmon', 'slateblue', 'sienna', 'violet', 'yellowgreen']
```

```python
# 设置作图主题
plt.style.use('fivethirtyeight') # plt.style.available可以查询可用的主题

# 初始化图表
fig, ax = plt.subplots(figsize=(10,5))

# 使用for循环作6家公司的股票收益曲线
for i in range(6):
    
    code = tickers[i] # 股票代码
    name = cn_name[i] # 股票中文名
    color = six_color[i] # 股票颜色
    last_value = df2.loc[end_date, code] # 股票最近一日的收盘价
    profit = "{0:+g}%".format(round(last_value - 100, 0)) # 计算收益率，整数百分比显示，正数前添加‘+’号
    
    # 作股价图
    ax.plot(df2[code], color = color, linewidth = 2)
    
    # 图上添加股票中文名及对应的收益率
    ax.text(x = '2019-08-31', y = last_value, s = name + "  " + profit, color = color, weight = 'bold')

# 添加底部水印
ax.text(x = '2018-01-01', y = 15, 
        s = '制图：MilanKaka              数据：Yahoo Finance', 
        fontsize = 10, color = 'grey')

ax.text(x = '2019-06-01', y = 15, 
        s = '设' + start_date + '股价为基准值100，数据截止 ' + end_date, 
        fontsize = 10, color = 'grey')

# 添加网格线作参照
ax.grid(alpha = 0.4)

# 添加图表标题
ax.set_title('中美互联网巨头2018年以来股票收益走势')

# 设定轴标签字体大小
ax.tick_params(labelsize=12)

# 设定x轴时间区间
ax.set_xlim('2018-01-01','2019-09-30')
```

新的图表更清晰地显示出了6家中美互联网公司2018年以来的股票涨跌走势。6家公司中只有亚马逊和谷歌的股票实现了正收益率，其余4家均是负收益率。3家美国公司股价总体表现大大好于3家中国公司，BAT均录得-10%以上的负收益。

此外结合贸易战的时间点，可以看到贸易战的几个重大时间点都影响了6家公司的股价，尤其是2018年3月贸易战开战、2019年8-9月新一轮关税加征和2019年5月特朗普推特宣布额外5000亿美元商品加征关税都重创了6家的股价。在贸易战有所缓和的2018年12月至2019年4月，除百度外都实现了不小的涨幅。

1. **亚马逊**：亚马逊在2018年是华尔街的宠儿，股价在2018年初随着利润指标改善即开始大幅增长，除受2018年3月贸易战开战影响到导致股价调整外，不断扩大涨幅，至2018年9月达到了70%的收益，成为历史第二支万亿美元市值公司，创始人贝索斯也因此荣登全球首富。此后随美股股指起落，自身增速放缓，收益率至今降到了49%。不过股神巴菲特今年都开始买入了，本就是基金重仓的亚马逊继续下跌也难。

2. **谷歌**：过于依赖广告收入的谷歌股价走势相对稳定，波幅在20%左右，在大部分的时间里维持在正收益率，不过广告业务的下滑、流量获取成本的上升，广告点击率所带来的收入下降、欧盟不断的反垄断罚款导致公司利润受影响，今年4月财报不及预期后一个月内跌去20%，好在7月的最新财报超预期，保住了一年多来9%的收益率。谷歌未来的股价仍然会依赖广告业务的表现。

3. **脸书**：2018年的脸书很失意，几次服务器宕机事件甚至都引起股价下跌，能维持-1%的收益率靠的是2019年的强势财报。2018年3月受总统选举数据泄露事件影响，收益率排在垫底位置，此后虽一路上涨，但是2018年半年报不及市场预期，公布当天暴跌近20%，不幸成为史上第一支单日市值下挫超千亿美元的个股，财报显示脸书的用户增长和营收增速放缓，股价也一直萎靡至2019年1月。此后的财报数据好转，至今股价勉强收复失地，好在Instagram用户快速增长，这款App有机会助推脸书的股价。

4. **阿里**：阿里的财报相对稳健，营收增速还能维持在40%以上，尤其核心的电商业务依然在狂奔，下沉市场挖掘战略显成效，月活用户仍能保持15%以上的增长。图上可以看到，在今年5月前阿里的股价走势和脸书在大部分时间里很接近（阿里波幅略小），今年5月特朗普发推加剧贸易战后，阿里跌幅超过脸书，2018年至今收益率为-10%，已经是BAT三家中表现最好的。在实体经济下滑、国家刺激消费扩大内需的情况下，电商业务为主的阿里应该还能保持不错的增速，有助于推动股价的增长。

5. **腾讯**：2018年12月前曾是6家里表现股价最差的公司，不仅受贸易战影响，还被大股东MIH减持，自身利润增速放慢，行业版号审批问题导致游戏业务大受影响，引发市场担忧，2019年更被港股整体市场拖累，最近的8月受恒生指数大跌牵连以及一份市场不认可的财报，股价跌去10%，最终录得-22%的收益率。亮点是金融科技和企业服务收入大增，最近两个财报季增速都超过44%。另一好消息是最近几个月获得版号的游戏增多，游戏业务开始回暖。下一财报季的表现值得期待。

6. **百度**：相信不仅是这6家中最让人失望的公司，也是有代表性的中概股中最让人失望的公司，-57%的收益率甚至差过年轻人的第一支腰斩股小米。主营的搜索业务还在不断被人诟病。2018年5月前百度曾经回光返照创造股价新高，5月陆奇离开，股价也跟着其他中概股一起一路下跌，但百度没有反弹。该来的总会来的，百度在2019年5月首次发布亏损的财报，期间至今又跌去40%，市值相继被美团、京东、拼多多超过。旧业务增长乏力、AI业务又无法在短时间内带来利润，中概股中有太多比百度更值得投资的标的。

综上来看，2018年来依赖广告业务的公司股价相对低迷，这与全球经济增长放缓有关，互联网红利的消退也导致互联网行业月活用户增长慢。贸易战也一直是笼罩在全球经济之上的乌云，特朗普时不时就发推打个雷。消费概念的亚马逊和阿里分别是美中互联网巨头里表现最好的股票，过去一年里A股表现最好的也是消费股（白酒、酱油、家电、机场等）。所以，仓位里配置一部分消费股对于散户来说是个不错的选择。当然，除百度外，上文的另外5家公司仍然是行业里的翘楚，用户粘性也高，在可预见的3-5年内也都是值得投资的标的。

## 附：
使用`plotly`绘制股票走势

```python
# 载入plotly
import plotly.graph_objects as go
```

```python
# 初始化图表
fig = go.Figure()

# 使用for循环分别画出股价走势
for i in range(6):
    
    code = tickers[i]
    name = cn_name[i]
    
    fig.add_trace(go.Scatter(x=df2.index, y=df2[code], name=name))

# 添加标题
fig.update_layout(title_text='中美互联网巨头2018年来股票收益走势')

# 设置x轴区间显示范围为1个月，6个月，年初至今，1年四个选项
fig.update_layout(
    xaxis=go.layout.XAxis(
        # 设置日期显示格式
        tickformat = '%Y-%m-%d%',
        
        # 设置x轴区间显示可选范围
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1月",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6月",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="年初至今",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1年",
                     step="year",
                     stepmode="backward"),
                dict(label='全部',
                     step="all")
            ])
        ),
        
        # 设置x轴区间滑块
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

fig.show()
```
